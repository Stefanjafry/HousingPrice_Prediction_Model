```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Zillow ZHVI dataset
df = pd.read_csv("ZHVI_data.csv")

# --- STEP 1: Basic Cleaning ---

# Drop rows with missing critical metadata
df_clean = df.dropna(subset=['RegionName', 'StateName']).copy()

# Identify columns containing time-series house price values (from column 6 onwards)
price_cols = df_clean.columns[6:]

# Forward-fill and backward-fill missing time series data safely
df_clean.loc[:, price_cols] = df_clean.loc[:, price_cols].ffill(axis=1).bfill(axis=1)

# --- STEP 2: Reshape Wide to Long Format ---

# Melt the DataFrame: each row = (Region, Date, Price)
df_long = df_clean.melt(
    id_vars=["RegionID", "SizeRank", "RegionName", "RegionType", "StateName"],
    var_name="Date",
    value_name="Price"
)

# Convert the 'Date' column to datetime format
df_long['Date'] = pd.to_datetime(df_long['Date'])

# Remove any remaining rows with null prices
df_long.dropna(subset=["Price"], inplace=True)

# --- STEP 3: Feature Engineering ---

# Add temporal features for analysis and modeling
df_long["Year"] = df_long["Date"].dt.year
df_long["Month"] = df_long["Date"].dt.month
df_long["Quarter"] = df_long["Date"].dt.to_period("Q")

# --- STEP 4: Exploratory Data Analysis (EDA) ---

# 1. Trend of average house price across all regions
plt.figure(figsize=(12, 5))
avg_trend = df_long.groupby("Date")["Price"].mean()
avg_trend.plot()
plt.title("Average U.S. House Price Over Time")
plt.xlabel("Date")
plt.ylabel("Average Price ($)")
plt.grid(True)
plt.tight_layout()
plt.show()

# 2. Latest month price distribution by state
latest_date = df_long["Date"].max()
latest_prices = df_long[df_long["Date"] == latest_date]

plt.figure(figsize=(14, 6))
sns.boxplot(data=latest_prices, x="StateName", y="Price")
plt.xticks(rotation=90)
plt.title(f"Distribution of House Prices by State as of {latest_date.date()}")
plt.ylabel("Price ($)")
plt.tight_layout()
plt.show()
```

Line Graph: 2000-2007 is the pre financial crisis growth, showing steady increase in average house prices which was driven by housing demand, loose credit, and speculative investments. During 2006-2012, the average house price starts to decline, which indicates the collapse of the housing bubble, mortgage defaults and the overall recession. Price decline is substancial approximiately 15-25% decline. From 2012 onwawads, average housing prices start increasing and eventually explode. The main drivers are low interest rates, the economic recovery from the financial crises and urban job growth.The explosive price appreciation, is particularly obsereved from the late 2020s through 2022 due to the pandemic. Remote work migration, low mortgage rate and inventory shortages are attributing factors to this explosion.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score,
    median_absolute_error
)
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.preprocessing import LabelEncoder

# Assume df_long is already cleaned and loaded
df_long["Quarter"] = df_long["Date"].dt.quarter

# --- Group-wise Lag Features ---
df_long = df_long.sort_values(['RegionName', 'Date']).copy()
df_long['Lag_1'] = df_long.groupby('RegionName')['Price'].shift(1)
df_long['Lag_3'] = df_long.groupby('RegionName')['Price'].shift(3)
df_long['Lag_6'] = df_long.groupby('RegionName')['Price'].shift(6)
df_long['Lag_12'] = df_long.groupby('RegionName')['Price'].shift(12)
df_long.dropna(inplace=True)

# Convert categorical features
df_long["Quarter"] = df_long["Quarter"].astype(int)
df_long['RegionName'] = df_long['RegionName'].astype('category')
df_long['StateName'] = df_long['StateName'].astype('category')
df_long['RegionType'] = df_long['RegionType'].astype('category')

# --- Feature Set ---
features = ['Lag_1', 'Lag_3', 'Lag_6', 'Lag_12', 'Year', 'Month', 'Quarter',
            'StateName', 'RegionName', 'RegionType', 'SizeRank']
X = df_long[features]
y = df_long['Price']

# --- Encoding for XGBoost (Label encoding only categorical features) ---
X_encoded = X.copy()
for col in ['RegionName', 'StateName', 'RegionType']:
    le = LabelEncoder()
    X_encoded[col] = le.fit_transform(X_encoded[col])

# --- Time-Based Train/Test Split ---
cutoff_date = df_long["Date"].quantile(0.8)
train_mask = df_long["Date"] <= cutoff_date
test_mask = df_long["Date"] > cutoff_date

X_train_lgb = X[train_mask]
X_test_lgb = X[test_mask]
X_train_xgb = X_encoded[train_mask]
X_test_xgb = X_encoded[test_mask]
y_train = y[train_mask]
y_test = y[test_mask]

# --- 1. LightGBM ---
lgb_model = LGBMRegressor(random_state=42)
lgb_model.fit(X_train_lgb, y_train)
lgb_pred = lgb_model.predict(X_test_lgb)

# --- 2. XGBoost ---
xgb_model = XGBRegressor(random_state=42, objective='reg:squarederror')
xgb_model.fit(X_train_xgb, y_train)
xgb_pred = xgb_model.predict(X_test_xgb)

# --- 3. SARIMAX (Only on top 5 largest regions for demo purposes) ---
top_regions = df_long['RegionName'].value_counts().head(5).index.tolist()
sarimax_mae, sarimax_rmse, sarimax_r2, sarimax_medae = [], [], [], []

for region in top_regions:
    region_df = df_long[df_long['RegionName'] == region].sort_values("Date")
    series = region_df.set_index("Date")['Price']
    train_series = series[series.index <= cutoff_date]
    test_series = series[series.index > cutoff_date]

    if len(train_series) < 36 or len(test_series) < 12:
        continue

    try:
        sarimax_model = SARIMAX(train_series, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
        sarimax_result = sarimax_model.fit(disp=False)
        sarimax_pred = sarimax_result.forecast(steps=len(test_series))

        sarimax_mae.append(mean_absolute_error(test_series, sarimax_pred))
        sarimax_rmse.append(np.sqrt(mean_squared_error(test_series, sarimax_pred)))
        sarimax_r2.append(r2_score(test_series, sarimax_pred))
        sarimax_medae.append(median_absolute_error(test_series, sarimax_pred))
    except:
        continue

# --- Evaluation ---
def evaluate_all(y_true, y_pred, label):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    smape = 100 * np.mean(2 * np.abs(y_pred - y_true) / (np.abs(y_pred) + np.abs(y_true)))
    r2 = r2_score(y_true, y_pred)
    medae = median_absolute_error(y_true, y_pred)

    print(f"\n {label} Performance:")
    print(f"MAE:     {mae:,.2f}")
    print(f"MedAE:   {medae:,.2f}")
    print(f"RMSE:    {rmse:,.2f}")
    print(f"MAPE:    {mape:.2f}%")
    print(f"SMAPE:   {smape:.2f}%")
    print(f"R:      {r2:.4f}")

# Run evaluations
evaluate_all(y_test, lgb_pred, "LightGBM (All Regions)")
evaluate_all(y_test, xgb_pred, "XGBoost (All Regions)")

# SARIMAX averaged over top regions
print("\n SARIMAX (Average over Top Regions):")
print(f"MAE:     {np.mean(sarimax_mae):,.2f}")
print(f"MedAE:   {np.mean(sarimax_medae):,.2f}")
print(f"RMSE:    {np.mean(sarimax_rmse):,.2f}")
print(f"R:      {np.mean(sarimax_r2):.4f}")
```

R = 0.9339 Explains ~93% of the variance in housing prices across all regions.
MAE under $10K means it's strong even at regional scale.
Slightly more accurate than XGBoost but with better stability (lower RMSE and SMAPE).

```python
from sklearn.model_selection import KFold
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score,
    median_absolute_error
)
import numpy as np
import pandas as pd

# Define evaluation metrics for a single fold
def evaluate_fold(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    medae = median_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    smape = 100 * np.mean(2 * np.abs(y_pred - y_true) / (np.abs(y_pred) + np.abs(y_true)))
    r2 = r2_score(y_true, y_pred)
    return [mae, medae, rmse, mape, smape, r2]

# Set up KFold
kf = KFold(n_splits=5, shuffle=True, random_state=42)
metrics = ['MAE', 'MedAE', 'RMSE', 'MAPE', 'SMAPE', 'R']

# --- LightGBM Cross-Validation ---
lgb_scores = []
lgb_model = LGBMRegressor(random_state=42)

for train_idx, test_idx in kf.split(X_encoded):
    X_train, X_test = X_encoded.iloc[train_idx], X_encoded.iloc[test_idx]
    y_train_cv, y_test_cv = y.iloc[train_idx], y.iloc[test_idx]
    lgb_model.fit(X_train, y_train_cv)
    y_pred = lgb_model.predict(X_test)
    lgb_scores.append(evaluate_fold(y_test_cv, y_pred))

# --- XGBoost Cross-Validation ---
xgb_scores = []
xgb_model = XGBRegressor(random_state=42, objective='reg:squarederror')

for train_idx, test_idx in kf.split(X_encoded):
    X_train, X_test = X_encoded.iloc[train_idx], X_encoded.iloc[test_idx]
    y_train_cv, y_test_cv = y.iloc[train_idx], y.iloc[test_idx]
    xgb_model.fit(X_train, y_train_cv)
    y_pred = xgb_model.predict(X_test)
    xgb_scores.append(evaluate_fold(y_test_cv, y_pred))

# Aggregate metrics
lgb_mean = np.mean(lgb_scores, axis=0)
lgb_std = np.std(lgb_scores, axis=0)
xgb_mean = np.mean(xgb_scores, axis=0)
xgb_std = np.std(xgb_scores, axis=0)

# Display results
cv_results = pd.DataFrame({
    'Model': ['LightGBM'] * 6 + ['XGBoost'] * 6,
    'Metric': metrics * 2,
    'Mean': np.concatenate([lgb_mean, xgb_mean]),
    'Std Dev': np.concatenate([lgb_std, xgb_std])
})

display(cv_results)
```

Performed 5-fold cross validation provides more reliable evaluation and dects overfitting.
In this scenario XG boost performed better.
For both models the MAE and RMSE improvements are significant. MAE and RMSE improvements are significant with MAPE/SMAPE under 0.5% which is increadibily accurate .Standard deviations are very low, suggesting high model stability across folds. For both models the R^2 is >0.999. This could imply overfitting.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import statsmodels.api as sm

# Use XGBoost predictions on test set from your earlier time-based split
residuals = y_test - xgb_pred
fitted = xgb_pred

# --- Studentized Residuals ---
influence = sm.OLS(residuals, sm.add_constant(fitted)).fit().get_influence()
studentized_residuals = influence.resid_studentized_internal

# --- Plot 1: Fitted vs Residuals ---
plt.figure(figsize=(12, 4))
sns.scatterplot(x=fitted, y=residuals)
plt.axhline(0, linestyle='--', color='red')
plt.title("Fitted vs. Residuals (XGBoost)")
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")
plt.grid(True)
plt.tight_layout()
plt.show()

# --- Plot 2: Studentized Residuals ---
plt.figure(figsize=(12, 4))
sns.histplot(studentized_residuals, bins=30, kde=True)
plt.axvline(0, linestyle='--', color='red')
plt.title("Studentized Residuals (XGBoost)")
plt.xlabel("Studentized Residuals")
plt.grid(True)
plt.tight_layout()
plt.show()

# --- Plot 3: Q-Q Plot ---
plt.figure(figsize=(6, 6))
sm.qqplot(residuals, line='45', fit=True)
plt.title("Q-Q Plot of Residuals (XGBoost)")
plt.tight_layout()
plt.show()
```

Fitted Values Vs Risiduals: There is a clear funnel pattern where the risiduals variance increase with higher fitted values. This means that model performace worsense in the later years. This suggests heteroscedasicity.
Student Residuals: It is a longer right skewed distribution, which indicates overpredictions. However overall it's still very compact, but mitigating this factor may help with model performance.
Q-Q plot: There is clear deviation at both tails of the Q-Q plot. The S shape indicates something called fat-tailed residuals. This is when extreme under and over predictions exist more than expected.

```python
from sklearn.linear_model import LinearRegression

# Prepare LightGBM input
X_train_lgb = X[train_mask]
X_test_lgb = X[test_mask]

# Train LightGBM
lgb_model = LGBMRegressor(random_state=42)
lgb_model.fit(X_train_lgb, y_train)
lgb_pred = lgb_model.predict(X_test_lgb)

# Residuals and fitted values
lgb_residuals = y_test - lgb_pred
lgb_fitted = lgb_pred

# Compute studentized residuals
influence_lgb = sm.OLS(lgb_residuals, sm.add_constant(lgb_fitted)).fit().get_influence()
studentized_residuals_lgb = influence_lgb.resid_studentized_internal

# Plot 1: Fitted vs. Residuals
plt.figure(figsize=(12, 4))
sns.scatterplot(x=lgb_fitted, y=lgb_residuals)
plt.axhline(0, linestyle='--', color='red')
plt.title("Fitted vs. Residuals (LightGBM)")
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot 2: Studentized Residuals Histogram
plt.figure(figsize=(12, 4))
sns.histplot(studentized_residuals_lgb, bins=30, kde=True)
plt.axvline(0, linestyle='--', color='red')
plt.title("Studentized Residuals (LightGBM)")
plt.xlabel("Studentized Residuals")
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot 3: Q-Q Plot
plt.figure(figsize=(6, 6))
sm.qqplot(lgb_residuals, line='45', fit=True)
plt.title("Q-Q Plot of Residuals (LightGBM)")
plt.tight_layout()
plt.show()
```

Studentized Residuals: More centered than XGBoost. LightGBM handles middle-range housing prices well.
Q-Q Plot Of Residuals: Better left tail, but extreme right tail sugessting extreme positive risudals.


To fix the high levels of heteroscedasicity and S curve Q-Q plots we will split the machine learning model, into low, mid and high tier models. This model fixes the 'one size fits all' mistake the previous models were making.

```python
import pandas as pd
import numpy as np

# Load the data
df = pd.read_csv("ZHVI_data.csv")

# Clean and preprocess
df_clean = df.dropna(subset=['RegionName', 'StateName']).copy()
price_cols = df_clean.columns[6:]
df_clean[price_cols] = df_clean[price_cols].ffill(axis=1).bfill(axis=1)

# Reshape to long format
df_long = df_clean.melt(
    id_vars=["RegionID", "SizeRank", "RegionName", "RegionType", "StateName"],
    var_name="Date",
    value_name="Price"
)
df_long['Date'] = pd.to_datetime(df_long['Date'], errors='coerce')
df_long.dropna(subset=["Price"], inplace=True)

# Add temporal features
df_long["Year"] = df_long["Date"].dt.year
df_long["Month"] = df_long["Date"].dt.month
df_long["Quarter"] = df_long["Date"].dt.quarter

# Add price tier using qcut
df_long['PriceTier'] = pd.qcut(df_long['Price'], q=3, labels=['Low', 'Mid', 'High'])

# Save segmented data
df_long.to_csv("ZHVI_segmented_by_price_tier.csv", index=False)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, median_absolute_error
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor
from sklearn.preprocessing import LabelEncoder

# Add lag features for each region
df_long = df_long.sort_values(['RegionName', 'Date']).copy()
for lag in [1, 3, 6, 12]:
    df_long[f'Lag_{lag}'] = df_long.groupby('RegionName')['Price'].shift(lag)
df_long.dropna(inplace=True)

# Encode categorical features
for col in ['RegionName', 'StateName', 'RegionType']:
    df_long[col] = df_long[col].astype('category')

# Prepare XGBoost-specific encoded dataset
X_encoded = df_long.copy()
for col in ['RegionName', 'StateName', 'RegionType']:
    le = LabelEncoder()
    X_encoded[col] = le.fit_transform(X_encoded[col])

# Features to use
features = ['Lag_1', 'Lag_3', 'Lag_6', 'Lag_12', 'Year', 'Month', 'Quarter',
            'StateName', 'RegionName', 'RegionType', 'SizeRank']

# Function to evaluate and return metrics
def evaluate_model(y_true, y_pred):
    return {
        'MAE': mean_absolute_error(y_true, y_pred),
        'MedAE': median_absolute_error(y_true, y_pred),
        'RMSE': np.sqrt(mean_squared_error(y_true, y_pred)),
        'MAPE': np.mean(np.abs((y_true - y_pred) / y_true)) * 100,
        'SMAPE': 100 * np.mean(2 * np.abs(y_pred - y_true) / (np.abs(y_pred) + np.abs(y_true))),
        'R2': r2_score(y_true, y_pred)
    }

# Train models for each tier
results = []
for tier in ['Low', 'Mid', 'High']:
    print(f"\n Training models for {tier} price tier")

    tier_df = df_long[df_long['PriceTier'] == tier]
    tier_X_lgb = tier_df[features]
    tier_y = tier_df['Price']

    tier_X_xgb = X_encoded.loc[tier_df.index, features]

    # Time-based split
    cutoff_date = tier_df["Date"].quantile(0.8)
    train_mask = tier_df["Date"] <= cutoff_date
    test_mask = tier_df["Date"] > cutoff_date

    X_train_lgb = tier_X_lgb[train_mask]
    X_test_lgb = tier_X_lgb[test_mask]
    X_train_xgb = tier_X_xgb[train_mask]
    X_test_xgb = tier_X_xgb[test_mask]
    y_train = tier_y[train_mask]
    y_test = tier_y[test_mask]

    # LightGBM
    lgb_model = LGBMRegressor(random_state=42)
    lgb_model.fit(X_train_lgb, y_train)
    lgb_pred = lgb_model.predict(X_test_lgb)
    lgb_metrics = evaluate_model(y_test, lgb_pred)

    # XGBoost
    xgb_model = XGBRegressor(random_state=42, objective='reg:squarederror')
    xgb_model.fit(X_train_xgb, y_train)
    xgb_pred = xgb_model.predict(X_test_xgb)
    xgb_metrics = evaluate_model(y_test, xgb_pred)

    results.append({
        'Tier': tier,
        'Model': 'LightGBM',
        **lgb_metrics
    })
    results.append({
        'Tier': tier,
        'Model': 'XGBoost',
        **xgb_metrics
    })

# Display results as DataFrame
results_df = pd.DataFrame(results)
display(results_df)
```

In order to fix our predictive issues with high tier values we Filter out all the prices that are in the midtier, but for some reason are still being included into the high tier

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score,
    median_absolute_error
)
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor

# --- 1. Load and Clean Data ---
df = pd.read_csv("ZHVI_data.csv")
df_clean = df.dropna(subset=['RegionName', 'StateName']).copy()
price_cols = df_clean.columns[6:]
df_clean[price_cols] = df_clean[price_cols].ffill(axis=1).bfill(axis=1)

# Reshape to long format
df_long = df_clean.melt(
    id_vars=["RegionID", "SizeRank", "RegionName", "RegionType", "StateName"],
    var_name="Date", value_name="Price"
)
df_long['Date'] = pd.to_datetime(df_long['Date'], errors='coerce')
df_long.dropna(subset=["Price"], inplace=True)

# Add temporal features
df_long["Year"] = df_long["Date"].dt.year
df_long["Month"] = df_long["Date"].dt.month
df_long["Quarter"] = df_long["Date"].dt.quarter

# --- 2. Recreate Price Tier ---
df_long['PriceTier'] = pd.qcut(df_long['Price'], q=3, labels=['Low', 'Mid', 'High'])

# --- 3. Filter to High Tier and Cap ---
high_tier_df = df_long[df_long['PriceTier'] == 'High'].copy()
price_ceiling = high_tier_df['Price'].quantile(0.95)
high_filtered_df = high_tier_df[high_tier_df['Price'] <= price_ceiling].copy()

# --- 4. Lag Features ---
high_filtered_df = high_filtered_df.sort_values(['RegionName', 'Date'])
for lag in [1, 3, 6, 12]:
    high_filtered_df[f'Lag_{lag}'] = high_filtered_df.groupby('RegionName')['Price'].shift(lag)
high_filtered_df.dropna(inplace=True)

# --- 5. Encode Categorical Features ---
for col in ['RegionName', 'StateName', 'RegionType']:
    high_filtered_df[col] = high_filtered_df[col].astype('category')

X_encoded = high_filtered_df.copy()
for col in ['RegionName', 'StateName', 'RegionType']:
    le = LabelEncoder()
    X_encoded[col] = le.fit_transform(X_encoded[col])

# --- 6. Feature and Target Setup ---
features = ['Lag_1', 'Lag_3', 'Lag_6', 'Lag_12', 'Year', 'Month', 'Quarter',
            'StateName', 'RegionName', 'RegionType', 'SizeRank']
X_lgb = high_filtered_df[features]
X_xgb = X_encoded[features]
y = high_filtered_df['Price']

# --- 7. Time-Based Train/Test Split ---
cutoff = high_filtered_df['Date'].quantile(0.8)
train_mask = high_filtered_df['Date'] <= cutoff
test_mask = high_filtered_df['Date'] > cutoff

X_train_lgb, X_test_lgb = X_lgb[train_mask], X_lgb[test_mask]
X_train_xgb, X_test_xgb = X_xgb[train_mask], X_xgb[test_mask]
y_train, y_test = y[train_mask], y[test_mask]

# --- 8. Train Models ---
lgb_model = LGBMRegressor(random_state=42)
lgb_model.fit(X_train_lgb, y_train)
lgb_pred = lgb_model.predict(X_test_lgb)

xgb_model = XGBRegressor(random_state=42, objective='reg:squarederror')
xgb_model.fit(X_train_xgb, y_train)
xgb_pred = xgb_model.predict(X_test_xgb)

# --- 9. Evaluation Function ---
def evaluate_model(y_true, y_pred, label):
    print(f"\n {label} Performance:")
    print(f"MAE:     {mean_absolute_error(y_true, y_pred):,.2f}")
    print(f"MedAE:   {median_absolute_error(y_true, y_pred):,.2f}")
    print(f"RMSE:    {np.sqrt(mean_squared_error(y_true, y_pred)):,.2f}")
    print(f"MAPE:    {np.mean(np.abs((y_true - y_pred) / y_true)) * 100:.2f}%")
    print(f"SMAPE:   {100 * np.mean(2 * np.abs(y_pred - y_true) / (np.abs(y_pred) + np.abs(y_true))):.2f}%")
    print(f"R:      {r2_score(y_true, y_pred):.4f}")

# --- 10. Evaluate Models ---
evaluate_model(y_test, lgb_pred, "LightGBM (High Tier Capped)")
evaluate_model(y_test, xgb_pred, "XGBoost (High Tier Capped)")
```

For the lightGBM risuals are not as tight as the XGBoost, however prediction error remains 10,000-15,000. For XGBoost the residuals are more tighter however when the housing prices are more recent perdiction jumps to the 10,000-20,000 range slightly more than LightGBM. 

Studentized residuals: Both models follow a bell-shaped distribution, which is idea. The peak is centered around zero, indicating the model is unbiased. LightGBM is more unbiased than XGBoost.

```python
import pandas as pd
import numpy as np
from dateutil.relativedelta import relativedelta
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor
from sklearn.preprocessing import LabelEncoder

# --- 1. Load the cleaned and segmented dataset ---
df_long = pd.read_csv("ZHVI_segmented_by_price_tier.csv")
df_long['Date'] = pd.to_datetime(df_long['Date'])

# --- 2. Create lag features ---
df_long = df_long.sort_values(['RegionName', 'Date'])
for lag in [1, 3, 6, 12]:
    df_long[f'Lag_{lag}'] = df_long.groupby(['RegionName', 'PriceTier'])['Price'].shift(lag)
df_long.dropna(inplace=True)

# --- 3. Encode categorical features for XGBoost ---
label_encoders = {}
df_encoded = df_long.copy()
for col in ['RegionName', 'StateName', 'RegionType', 'PriceTier']:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col])
    label_encoders[col] = le

# --- 4. Set up forecasting loop ---
forecast_results = []
features = ['Lag_1', 'Lag_3', 'Lag_6', 'Lag_12', 'Year', 'Month', 'Quarter',
            'StateName', 'RegionName', 'RegionType', 'SizeRank']

for tier in ['Low', 'Mid', 'High']:
    tier_df = df_long[df_long['PriceTier'] == tier].copy()
    encoded_df = df_encoded.loc[tier_df.index]

    # Train/Test Split
    cutoff = tier_df['Date'].quantile(0.8)
    train_mask = tier_df['Date'] <= cutoff
    X_train_lgb = tier_df[features][train_mask].copy()
    for col in ['StateName', 'RegionName', 'RegionType']:
        X_train_lgb[col] = X_train_lgb[col].astype('category')
    y_train = tier_df['Price'][train_mask]
    X_train_xgb = encoded_df[features][train_mask]

    # Train models
    lgb_model = LGBMRegressor(random_state=42)
    lgb_model.fit(X_train_lgb, y_train)

    xgb_model = XGBRegressor(random_state=42, objective='reg:squarederror')
    xgb_model.fit(X_train_xgb, y_train)

    # --- 5. Forecast for each region ---
    last_date = tier_df['Date'].max()
    regions = tier_df['RegionName'].unique()

    for region in regions:
        region_df = tier_df[tier_df['RegionName'] == region].copy().sort_values("Date")

        if len(region_df) < 12:
            continue  # not enough history

        history = region_df.copy()
        state = history["StateName"].iloc[0]
        region_type = history["RegionType"].iloc[0]
        size_rank = history["SizeRank"].iloc[0]

        for i in range(1, 7):  # next 6 months
            next_date = last_date + relativedelta(months=i)
            year, month = next_date.year, next_date.month
            quarter = (month - 1) // 3 + 1

            lags = {}
            for lag in [1, 3, 6, 12]:
                if len(history) >= lag:
                    lags[f'Lag_{lag}'] = history['Price'].iloc[-lag]
                else:
                    lags[f'Lag_{lag}'] = np.nan

            if any(pd.isna(val) for val in lags.values()):
                break

            # Prepare LightGBM input
            row_lgb = pd.DataFrame({
                'Lag_1': [lags['Lag_1']],
                'Lag_3': [lags['Lag_3']],
                'Lag_6': [lags['Lag_6']],
                'Lag_12': [lags['Lag_12']],
                'Year': [year],
                'Month': [month],
                'Quarter': [quarter],
                'StateName': [state],
                'RegionName': [region],
                'RegionType': [region_type],
                'SizeRank': [size_rank]
            })

            for col in ['StateName', 'RegionName', 'RegionType']:
                row_lgb[col] = row_lgb[col].astype('category')

            lgb_pred = lgb_model.predict(row_lgb)[0]

            # Prepare XGBoost input
            row_xgb = row_lgb.copy()
            for col in ['RegionName', 'StateName', 'RegionType']:
                row_xgb[col] = label_encoders[col].transform(row_xgb[col])

            xgb_pred = xgb_model.predict(row_xgb)[0]

            # Store prediction
            forecast_results.append({
                'PriceTier': tier,
                'RegionName': region,
                'Date': next_date,
                'LightGBM_Prediction': lgb_pred,
                'XGBoost_Prediction': xgb_pred
            })

            # Update history with LGBM prediction for next lags
            history = pd.concat([
                history,
                pd.DataFrame([{'Date': next_date, 'Price': lgb_pred}])
            ], ignore_index=True)

# --- 6. Save forecast ---
forecast_df = pd.DataFrame(forecast_results)
forecast_df.to_csv("6_month_price_tier_forecasts.csv", index=False)
print(" Forecasts saved to '6_month_price_tier_forecasts.csv'")
```
